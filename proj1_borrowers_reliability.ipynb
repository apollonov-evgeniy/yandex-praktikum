### Лемматизация

```python
from pymystem3 import Mystem
m = Mystem()

#------ДОПОЛНЕНИЕ------
#создадим строку с лемматезированными значениями
string_lemm=' '.join(data['purpose'])
lemm_2=m.lemmatize(string_lemm)  
from collections import Counter
print(Counter(lemm_2).most_common(20))
#------------------------------------------
#выберем  'недвижимость','жилье', 'автомобиль', 'образование', 'свадьба'
#остальные не подходят
#--------------------------------------------

#создаем список уникальных значений столбца purpose' и лемматизируем их 
lemmas=[]
purpose_unique=data['purpose'].unique()
for purpose_unique_str in purpose_unique: 
        lemmas.append(m.lemmatize(purpose_unique_str))

#display(lemmas) #вывод на экран для создания словаря

#вручную создаем словарь целей кредита для data['purpose'] 
lemmas_dict=pd.Series(['недвижимость','жилье', 'автомобиль', 'образование', 'свадьба'])

#функция выдает значение из словаря lemmas_dict для каждого значения столбца 'purpose'

def lem(text): #функция замены текста на значения словаря
  
    lemm=m.lemmatize(text)
    for str in lemmas_dict:
        if str in lemm:
            if str == 'жилье':# замена жилье на недвижимость
                return 'недвижимость'
            else:    
                return str
        
data['purpose']=data['purpose'].apply(lem) #замена значения 'purpose' на значения словаря

#display(data.head(10))
```

### Вывод

При лемматизации был создан список lemmas с уникальными значениями столбца 'purpose', из которого были взяты значения обобщенной цели кредита: 'недвижимость','жилье', 'автомобиль', 'образование', 'свадьба'.Для категоризации. Стоит обратить внимание на значения 'недвижимость' и 'жилье, которые по смыслу являются синонимами. функция lem заменяет значения в изначальном столбце на три категории.
